```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# composicións
x = np.linspace(0,1)
T = 273.15 + 20 # Celsius

# Antoine benceno
A_b = 4.01814
B_b = 1203.835
C_b = -53.226
# Antoine tolueno
A_t = 4.08245
B_t = 1346.382
C_t = -53.508
```

```python
def P_sat_b(T):
    return 10 ** ( A_b - B_b / ( T + C_b ) )
```

```python
def P_sat_t(T):
    return 10 ** ( A_t - B_t / ( T + C_t ) )
```

```python
# Diagrama Pxy

P = x * P_sat_b(T) + ( 1 - x ) * P_sat_t(T)
y = x * P_sat_b(T) / ( x * P_sat_b(T) + ( 1 - x ) * P_sat_t(T) )
```

```python
plt.plot(x,P,y,P)
plt.grid(visible='yes', which='both', axis='both')
plt.xlim([0,1])
plt.legend(['Curva de líquido', 'Cura de vapor'])
```
